```python
## Exploratory Data Analysis (EDA)

# 1. Missing Values Analysis
print("="*50)
print("MISSING VALUES ANALYSIS")
print("="*50)
missing_data = df.isnull().sum()
missing_percent = (df.isnull().sum() / len(df)) * 100
missing_df = pd.DataFrame({'Missing_Count': missing_data, 'Percentage': missing_percent})
print(missing_df[missing_df['Missing_Count'] > 0])
print("\n")
```

```python
# 2. Statistical Summary of Numeric Columns
print("="*50)
print("STATISTICAL SUMMARY - NUMERIC COLUMNS")
print("="*50)
print(df.describe().round(2))
print("\n")
```

```python
# 3. Categorical Data Analysis
print("="*50)
print("CATEGORICAL DATA ANALYSIS")
print("="*50)
categorical_cols = ['Status', 'Type_Main', 'Manufacturer']
for col in categorical_cols:
    print(f"\n{col} - Value Counts:")
    print(df[col].value_counts())
```

```python
# 4. Correlation Analysis
print("="*50)
print("CORRELATION ANALYSIS")
print("="*50)
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
correlation_matrix = df[numeric_cols].corr()
print(correlation_matrix.round(3))
```

```python
# 5. Correlation Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', center=0)
plt.title('Correlation Heatmap - Numeric Variables')
plt.tight_layout()
plt.show()
```

```python
# 6. Distribution of Numeric Variables
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
fig.suptitle('Distribution of Numeric Variables', fontsize=16, fontweight='bold')

numeric_columns = ['year_introduced', 'speed_mph', 'height_ft', 'Inversions', 'G-force', 'latitude']
for idx, col in enumerate(numeric_columns):
    row = idx // 3
    col_pos = idx % 3
    # Handle NaN values
    data_clean = df[col].dropna()
    axes[row, col_pos].hist(data_clean, bins=30, edgecolor='black', alpha=0.7)
    axes[row, col_pos].set_title(f'{col}')
    axes[row, col_pos].set_xlabel(col)
    axes[row, col_pos].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# 7. Categorical Variables Visualization
fig, axes = plt.subplots(1, 3, figsize=(16, 5))
fig.suptitle('Categorical Variables Distribution', fontsize=16, fontweight='bold')

categorical_cols = ['Status', 'Type_Main', 'Manufacturer']
for idx, col in enumerate(categorical_cols):
    top_values = df[col].value_counts().head(10)
    axes[idx].barh(range(len(top_values)), top_values.values)
    axes[idx].set_yticks(range(len(top_values)))
    axes[idx].set_yticklabels(top_values.index)
    axes[idx].set_title(f'{col} (Top 10)')
    axes[idx].set_xlabel('Count')
    axes[idx].invert_yaxis()

plt.tight_layout()
plt.show()
```

```python
# 8. Box Plots for Numeric Variables
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
fig.suptitle('Box Plots - Numeric Variables', fontsize=16, fontweight='bold')

numeric_columns = ['year_introduced', 'speed_mph', 'height_ft', 'Inversions', 'G-force', 'latitude']
for idx, col in enumerate(numeric_columns):
    row = idx // 3
    col_pos = idx % 3
    data_clean = df[col].dropna()
    axes[row, col_pos].boxplot(data_clean)
    axes[row, col_pos].set_title(f'{col}')
    axes[row, col_pos].set_ylabel('Value')

plt.tight_layout()
plt.show()
```

```python
# 9. Coaster Introduction Trends Over Time
fig, ax = plt.subplots(figsize=(14, 6))
coasters_per_year = df['year_introduced'].value_counts().sort_index()
ax.plot(coasters_per_year.index, coasters_per_year.values, marker='o', linewidth=2, markersize=4)
ax.set_title('Number of Coasters Introduced Over Time', fontsize=14, fontweight='bold')
ax.set_xlabel('Year')
ax.set_ylabel('Number of Coasters')
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print("\nCoasters per Year Summary:")
print(coasters_per_year.describe())
```
